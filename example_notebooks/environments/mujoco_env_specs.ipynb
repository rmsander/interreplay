# Action and State Spaces of MuJuCo Gym Environments
This notebook can be used to explore specs of different OpenAI Gym MuJoCo environments, such as the dimensions of their observation, action, and reward spaces.

## Import Block

```python
import numpy as np
import gym
import mujoco_py
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import PIL
from IPython import display
```

## Example: Run HalfCheetah-v2
The loop below runs code for the Gym MuJoCo `HalfCheetah-v2` environment.

```python
# Number of steps to take in environment
N_steps = 50 

# Show sample trajectory in Cheetah environment
env_c = gym.make('HalfCheetah-v2')
env_c.reset()
A = env_c.render(mode='rgb_array').astype(np.uint8)
img = plt.imshow(A) # only call this once
D = np.zeros(A.shape + (N_steps,)).astype(np.uint8)
for i in range(N_steps):
    
    A = env_c.render(mode='rgb_array')
    D[..., i] = A
    img.set_data(A) # just update the data
    display.display(plt.gcf())
    display.clear_output(wait=True)
    action = env_c.action_space.sample()
    env_c.step(action)
```

## Example: Run InvertedPendulum-v2
The loop below runs code for the Gym MuJoCo `InvertedPendulum-v2` environment.

```python
N_steps = 50

# Show sample trajectory in inverted pendulum environment
env_i = gym.make('InvertedPendulum-v2')
env_i.reset()
A = env_i.render(mode='rgb_array').astype(np.uint8)
img = plt.imshow(A) # only call this once
D = np.zeros(A.shape + (N_steps,)).astype(np.uint8)
for i in range(N_steps):
    
    A = env_i.render(mode='rgb_array')
    D[..., i] = A
    img.set_data(A) # just update the data
    display.display(plt.gcf())
    display.clear_output(wait=True)
    action = env_i.action_space.sample()
    env_i.step(action)
```

## `HalfCheetah-v2`
Here, we find that:
1. **State Space dimensionality** = 17
2. **Action Space dimensionality** = 6

## Inspecting the Observation Space
We can also use the OpenAI Gym `mujoco-py` API to inspect the dimensions of data sources for these different environments. 

### Example: Inspecting HalfCheetah-v2 Environment|

```python
env_cheetah = gym.make('HalfCheetah-v2')
action_space_cheetah = env_cheetah.action_space
state_space_cheetah = env_cheetah.observation_space
print("Observation space: {} \nAction space: {} \n".format(state_space_cheetah, action_space_cheetah))
print("Observation Sample: \n{} \n".format(env_cheetah.reset()))
print("Action Sample: \n{} \n".format(action_space_cheetah.sample()))
s_t1 = env_cheetah.step(action_space_cheetah.sample())
print("Output: \n{}".format(s_t1))
```
